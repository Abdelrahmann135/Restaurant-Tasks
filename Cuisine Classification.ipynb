```python
import pandas as pd
dataset = pd.read_csv("./Dataset .csv")
dataset.drop_duplicates(subset=["Cuisines"])
dataset.head()
```

```python
lowerlimit = dataset['Average Cost for two'].quantile(0.05)
upperlimit = dataset['Average Cost for two'].quantile(0.95)
dataset = dataset[(dataset['Average Cost for two'] > lowerlimit) & (dataset['Average Cost for two'] < upperlimit)]
```

```python
dataset.isnull().sum()
dataset = dataset.dropna()
```

```python
x = dataset.drop(["Restaurant ID", "Address","Rating color", "Locality", "Cuisines"], axis=1)
y = pd.DataFrame(dataset["Cuisines"])
y
```

```python
x = pd.get_dummies(x, drop_first=True)
# y = pd.get_dummies(y, drop_first=True)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y['Cuisines'] = le.fit_transform(y['Cuisines'])
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
y_train = y_train.to_numpy().ravel()  
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='micro')
recall = recall_score(y_test, y_pred, average='micro')
f1 = f1_score(y_test, y_pred, average='micro')
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")
```

```python
from sklearn.linear_model import LogisticRegression
y_train = y_train.ravel()  
model2 = LogisticRegression(max_iter=500)
model2.fit(x_train, y_train)
```

```python
y_pred2 = model2.predict(x_test)
accuracy = accuracy_score(y_test, y_pred2)
precision = precision_score(y_test, y_pred2, average='micro')
recall = recall_score(y_test, y_pred2, average='micro')
f1 = f1_score(y_test, y_pred2, average='micro')
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")
```
