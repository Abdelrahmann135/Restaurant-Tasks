```python
import pandas as pd
dataset = pd.read_csv("./Dataset .csv")
```

```python
data = dataset[['Restaurant ID','Restaurant Name','Cuisines', 'Average Cost for two','Aggregate rating','Votes']]
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
data['Restaurant Name'].value_counts()
```

```python
data['Restaurant Name'].duplicated().sum()
```

```python
data['Restaurant Name'].value_counts()
```

```python
data = data.sort_values(by=['Restaurant Name','Aggregate rating'],ascending=False)
data.head()
```

```python
data = data.drop_duplicates('Restaurant Name',keep='first')
data.head()
```

```python
data['Restaurant Name'].value_counts()
```

```python
data['Cuisines'] = data['Cuisines'].str.split(', ')
data.head()
```

```python
data = data.explode('Cuisines')
data.head()
```

```python
data['Cuisines'].value_counts()
```

```python
XRestCuis = pd.crosstab(data['Restaurant Name'], data['Cuisines'])
XRestCuis.head()
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(XRestCuis.loc["Pita Pit"].values,XRestCuis.loc["Ooma"].values)
```

```python
from scipy.spatial.distance import pdist, squareform
jaccardDist = pdist(XRestCuis.values, metric='jaccard')
jaccardMatrix = squareform(jaccardDist)
jaccardSim = 1 - jaccardMatrix
dfJaccard = pd.DataFrame(jaccardSim,index=XRestCuis.index,columns=XRestCuis.index)
dfJaccard.head()
```

```python
resto_reco = 'Maori'
sim = dfJaccard.loc[resto_reco].sort_values(ascending=False)
sim = pd.DataFrame({'Restaurant Name': sim.index, 'simScore': sim.values})
sim = sim[(sim['Restaurant Name']!= resto_reco) & (sim['simScore']>=0.7)].head()
RestoRec = pd.merge(sim,data[['Restaurant Name','Aggregate rating']],how='inner',on='Restaurant Name')
FinalRestoRec = RestoRec.sort_values('Aggregate rating',ascending=False).drop_duplicates('Restaurant Name',keep='first')
FinalRestoRec
```
